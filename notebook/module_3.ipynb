```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 128

times = np.arange(0, N, 1)

freq = 21/128.0
x = np.cos(2*np.pi*times*freq)
X = np.fft.fft(x)
```

```python
plot(x);
show();
stem(abs(X));
```

Observemos que en este caso, la frecuencia es un múltiplo de la frecuencia fundamental, con lo cuál tiene sentido que haya dos picos exactos en el espectograma.

Hagamos lo mismo pero con $f_0 = 21/127$. En este caso la frecuencia no es un múltiplo...

```python
freq = 21/127.0
x = np.cos(2*np.pi*times*freq)
X = np.fft.fft(x)
```

```python
plot(x);
show();
stem(abs(X));
```

En este caso, como no es un múltiplo exacto, la potencia se "distribuye" entre varios componentes.
